# Machine learning - Branches

 * **Supervised**
    + Map a given input (sample) to a given output (class/value)
    + Supervised, because we during training always knows what the expected output should be.
    + **Most common**
    + Two major variants: 
       - Regression (predict value)
       - Classification (predict class)

 * **Self-supervised**
    + Special case of supervised learning - **no human annotated data**
    + Examples:
        - Autoencoders: encode information and then decode to the same input, trying to learn what is important about the data
        - Word embeddings: predict missing word from a context

# Machine learning - Branches

 * **Unsupervised**
    + Find patterns in the data, denoising
    + E.g. dimensionality reduction and clustering

 * **Reinforcement learning**
    + Modelled around an **agent** in a world/**environement**
    + The agent recieves information about the enviroment and take **actions**
    + Learns by maximizing a **reward** from these taken actions.

# Machine learning (ML) - High level workflow
 * Find a ML problem: Define a **task**
   + Define **Input/Output**: What  goes into and out of the system

 * Choose a **measure of success**
   + Answer the question: how can we objectively quantify how well our model works?
 * Train/Evaluation protocol
   + What data do we train on, what data do we evaluate on?

 * Construct a **model** - construct a neural network that may solve our problem

# Machine learning (ML) - High level workflow

 * **Train** the model on your data i.e. fit (find parameters for) our model

 * **Evaluate** the model
   + Evaluate it against a dataset the model has never seen to asses the models ability to generialize
   + May contain an error analysis step

 * **Adjust**/**Improve** model and/or hyperparameters, and repeat to satisfaction.
   + From objective metrics, adjust different parameters and retrain the model too se what effect this parameters have.

 * **Deploy**/**Reuse** in an application or use in another project

A pixel can have one or more **channels**. Color images has three channels: red, green and blue.

Intensitites are **quantized** to a integer, meaning there are a discrete number of possible intensities.
 
The number of discrete steps is called bit depth. Common bit depths:

 * 8 bit: 256 steps
 * 12 bit: 4 096 steps
 * 14 bit: 16 384 steps
 * 16 bit: 65 536 steps

# Images - Modes

Images can be divided into 2 broad categories:
 
 * Single channel:
   -  **Grayscale**, Luminosity - 1 channel per pixel
 * Multiple channels: 
   - **Color** - 3 channels per pixel, represents intensity of the individual channels: Red, Green and Blue
   - **Generalized** - N channels per pixel: different staining, filters, or frequencies (visible light, infrared, X-ray, etc)
      + More than 4 channels are typically not supported by common image libraries. The solution is to split these multichannel images into multiple single channel and combine before training.

# Normalization

Images are normally in integer formats, meaning they are quantized and all computations are exact, but cannot handle fractions

 * Neural networks **perform better** when images are **normalized**: 0 to 1, or -1 to 1
     + Reasons: how activation functions work and possible numerical stability issues with larger weights/inputs

# Normalization

### Min-Max normalization

Strech the image out from 0 to 1

````python
io.imshow((im - np.min(im)) / (np.max(im) - np.min(im)))
````

# Normalization

### Percentile normalization

````python
# normalize to [0,1]
percentile=99.9

high = np.percentile(im, percentile)
low = np.percentile(im, 100-percentile)

print("low", low, "high", high)

img = np.minimum(high, im)
img = np.maximum(low, img)

io.imshow( (img - low) / (high - low) ) # gives float64
````

# Normalization

# Loss functions
### Minimization objective when traing models

  + Used to guide the optimizer in how to change the weights to lower the loss value
  + **Crossentropy** - Negative Log Likelihood
    - Default **classification** loss
    - Binary or Categorical: true/false or categories (classes)
    - Special variant is Sparse categorical crossentropy (it expects a single class as an integer)
    - Fast convergence is a key property
  + Mean Squared Error (MSE)
    - Good default loss for **regression**
    - Yields large loss values
  + Mean Absolute Error (MAE)
    - Regression error, typically converges slower than MSE
    - Easier to interpret as it is the mean difference between true and predicted
  + Cosine Similarity
    - Commonly used for relevance and ranking problems
    - Similarity metric - vectors pointing in the same direction has 1, -1 is in the opposite direction, 0 is orthogonal to each other
  + Hinge loss or maximum margin - used in SVM

# Task objective

  + **Accuracy**
     - Percentage of samples that are predicted correctly
  + **F1**
     - Recall: finding relevant samples 
     - Precion: of those found relevant samples that are actually relevant
     - You can easily get good recall, find all relevant samples (overpredict at the cost of many errors)
     - You can easily get good precision, find very few relevant samples but those that we find are correct
     - You want a balance between recall and precision => F1
  + **MCC** (Matthews correlation coefficent) is a better metric of accuracy in binary classification today
     - Normalized between -1 and 1, where -1 is total disagreement, 0 is random prediction, 1 is perfect prediction

  + **ROC**
    - F1 gives a single number, but ROC gives a plot for different thresholds

# Tensorflow data pipeline - Reading images

````python
def load_image(image_path, y):
    data = tf.io.read_file(image_path)
    
    im = tf.io.decode_png(data, dtype=tf.dtypes.uint16)
    
    im_float = tf.image.convert_image_dtype(im, dtype=tf.float32)
    #im_float = tf.image.random_brightness(im_float, 0.2) + 0.2
    return im_float, y
````

````python
X_images = tf.constant(sorted(glob.glob("images/*.png")))  # Image paths
y_images = tf.constant(np.arange(0,200))                   # Id, label or other data

dataset = tf.data.Dataset.from_tensor_slices( (X_images,y_images) )\
                         .map(load_image, num_parallel_calls=tf.data.AUTOTUNE)
````

# Neural networks
 * A full network, or a model is represented in Tensorflow Keras as an instance of ``Model``

<img src="slide_images/nn.svg" width="50%"/>

 * Inspired by neurons in our brains, and converted into a mathematical model
 * These models are not trying to be accurate to how our brains work - but was the source for inspiration
 * In Machine Learning, neural networks can be described as a sequence of transformations from an input (observation) to a desired output (label/representation).

# Neural network - Models and Layers

### Models:
 * Defines input and output
 * **Encapsulates the entire network**
 * Provides prediction and train methods
 * Is *not restricted to a sequence*, is in fact a graph of operations, multiple paths that converges is possible.
 * *Multiple* input and outputs are supported

### Layers:

 * Parametrical functions / functions with weights
 * Initializers for weights, typically random
 * Represents a single transformation
 * **Building block**


# Neural networks

Can be subdivided into multiple types, referring to properties of its architecture:
    
  * Artificial Neural Networks (ANN) - input, number of hidden layers and an output layer
  * Convolutional Neural Networks (CNN) - A neural network with convolutional layers
  * Recurrent Neural Networks (RNN) - A neural network with feedback loops (networks that operate on sequences of data)

# Building blocks - Activation functions

 * Introduces non-linearity into the models - can fit to more complex inputs, such found in the real world

````python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12,4))
fig.suptitle("Activation functions")

# ReLU
ax1.title.set_text("ReLU")
ax1.plot(np.linspace(-2, 2, 100), activations.relu(np.linspace(-2, 2, 100)).numpy())
ax1.set_xlim([-2, 2])
ax1.set_ylim([-2, 2])
ax1.grid()

# Sigmoid
ax2.title.set_text("Sigmoid")
ax2.plot(np.linspace(-2,2, 100), activations.sigmoid(np.linspace(-2, 2, 100)).numpy())
ax2.set_xlim([-2, 2])
ax2.set_ylim([-2, 2])
ax2.grid()

# Tanh
ax3.title.set_text("Tanh")
ax3.plot(np.linspace(-2, 2, 100), activations.tanh(np.linspace(-2, 2, 100)).numpy())
ax3.set_xlim([-2, 2])
ax3.set_ylim([-2, 2])
ax3.grid()
````

 * **ReLU** - Can not be negative, unbounded, can be stuck at 0
 * **Sigmoid** - Goes towards 0 in negative direction, 1 in positve direction
 * **Tanh** - Goes towards -1 and 1

# Building blocks - Activation functions contd.

Two common functions with special properties:

 * **Sigmoid** - typically used as the final activation, gives 1 output as a probability 0 to 1.
 * **Softmax** - typically used as the final activation for categorical classification, many classes. All activations sum to 1.
 * **ReLU** (only positive - to infinity) or **Linear** (postive and negative) - used for regression

````python
softmax = layers.Softmax()
````

````python
softmax([5., 1., 0.]).numpy()
````

````python
softmax([1., 0.2, 0.]).numpy()
````

One property of softmax is that it is scale sensitive, even with same ratio of input values

# Building blocks - Convolutional layers

Convolutional layers have 5 primary parameters:
 * **Filters**      - Number of outputs, also number of kernels
 * **Kernel size**  - The size of a kernel, width and height of the weight matrix
 * **Dilation**     - Spreading of a kernel when applied
 * **Stride**       - Step size, higher step sizes skips pixels in between
 * **Padding**      - SAME or VALID in tensorflow, controls how the border cases should be handled
